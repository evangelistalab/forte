```python
import nbimporter
import succ
import forte
import os

forte.startup()
```

```python
def make_geom(n,r):
    """Generate xyz coordinates for a linear chain of n hydrogen atoms with H-H distance equal to r"""
    geom = []
    for i in range(n):
        geom.append(f'H 0.0 0.0 {r * i:.4f}')
    return '\n'.join(geom)

ordering = 4
geom = make_geom(4,1.5)
psi4_wfn, energies = succ.run_psi(geom)
forte_objs = succ.make_forte_objs(psi4_wfn,mo_spaces={})

results = []
calc_data = succ.run_selected_ucc(forte_objs,psi4_wfn, 2, 0.0,ordering)  
```
